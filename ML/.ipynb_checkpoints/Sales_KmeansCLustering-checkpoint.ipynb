```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('sales_data_sample.csv', encoding="ISO-8859-1")
```

```python
data.head()
```

```python
columns = data.columns
```

```python
numeric_data = data.select_dtypes(include=[np.number]).dropna()
```

```python
scaler = StandardScaler()
#0 mean unit standard deviation
scaled_data = scaler.fit_transform(numeric_data)
```

```python
#Selecting optimal clusters 

wcss = []

k_range = range(1,11)

for k in k_range:
    kmeans = KMeans(n_clusters=k,random_state=0)
#     kmeans model fitted on scaled data
    kmeans.fit(scaled_data)
#     wcss computed using inertia_ attribute of kmeans
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_range,wcss,marker='o',linestyle='-',color = 'b')
plt.title('Elbow Plot ')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.grid(True)
plt.show()
```

```python
optimal_k = 3

kmeans = KMeans(n_clusters=optimal_k,random_state=0)
kmeans.fit(scaled_data)

data['Cluster'] = kmeans.labels_ 

plt.figure(figsize=(8, 5))
plt.scatter(numeric_data['SALES'],numeric_data['PRICEEACH'],c=data['Cluster'],cmap='plasma')
plt.xlabel("Sales")
plt.ylabel("PRICEEACH")
plt.show()

```
